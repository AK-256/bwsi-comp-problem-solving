# Intermediate Python: Class Inheritance

## Review of Object Oriented Programming

### What is OOP?

**Object-Oriented Programming (OOP)** organizes code around objects that contain attributes and methods.

## Inheritance

**Inheritance** allows a class to inherit attributes and methods from another class.

## Example: Dog Breeds

```python
class Dog:
    """A general Dog class"""

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        return f"{self.name} says: Woof!"

    def get_info(self):
        return f"{self.name} is {self.age} years old"
```

```python
generic_dog = Dog("Rex", 5)

print(generic_dog.bark())
print(generic_dog.get_info())
```

### Child Classes

```python
class GoldenRetriever(Dog):
    """Friendly dog that loves water"""

    def __init__(self, name, age, loves_swimming=True):
        super().__init__(name, age)
        self.loves_swimming = loves_swimming

    def fetch(self):
        return f"{self.name} retrieves the ball!"

    def bark(self):
        return f"{self.name} says: Woof woof! (friendly bark)"


class Poodle(Dog):
    """Intelligent and elegant dog"""

    def __init__(self, name, age, haircut_style="standard"):
        super().__init__(name, age)
        self.haircut_style = haircut_style

    def perform_trick(self):
        return f"{self.name} does a fancy trick!"

    def bark(self):
        return f"{self.name} says: Yip yip! (high-pitched bark)"
```

```python
buddy = GoldenRetriever("Buddy", 3)
princess = Poodle("Princess", 2, haircut_style="teddy bear")

print("=== Golden Retriever ===")
print(buddy.bark())
print(buddy.fetch())
print(buddy.get_info())
print(f"Loves swimming: {buddy.loves_swimming}")

print("\n=== Poodle ===")
print(princess.bark())
print(princess.perform_trick())
print(princess.get_info())
print(f"Haircut style: {princess.haircut_style}")
```

## Your Turn: Cat Exercise

```python
class Cat:
    """A general Cat class"""

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def meow(self):
        return f"{self.name} says: Meow!"

    def get_info(self):
        return f"{self.name} is {self.age} years old"
```

### Task 1: Create a Tuxedo Child Class

```python
class TuxedoCat(Cat):
    """A Tuxedo Cat"""

    def __init__(self, name, age, has_bow_tie=True):
        super().__init__(name, age)
        self.has_bow_tie = has_bow_tie

    def meow(self):
        return f"{self.name} says: meowmeow i hate tuxedos meowmeow!"

    def look_fancy(self):
        return f"{self.name} struts on two legs and slicks their whiskers back!"

    def get_info(self):
        return f"{self.name} is {self.age} years old. Bow tie: {self.has_bow_tie}"
```

### Task 2: Create Instances

```python
whiskers = Cat("Whiskers", 4)
mr_mittens = TuxedoCat("Mr. Mittens", 3, True)

print("=== Generic Cat ===")
print(whiskers.meow())
print(whiskers.get_info())

print("\n=== Tuxedo Cat ===")
print(mr_mittens.meow())
print(mr_mittens.get_info())
print(mr_mittens.look_fancy())
print(f"Has bow tie: {mr_mittens.has_bow_tie}")
```
