## Review of Object Oriented Programming

### What is OOP?

**Object-Oriented Programming (OOP)** is a programming paradigm that organizes code around **objects** — bundles of data (attributes) and behaviors (methods) that work together.

## Your Turn: Cat Exercise

```python
class Cat:
    """A general Cat class"""

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def meow(self):
        return f"{self.name} says: Meow!"

    def get_info(self):
        return f"{self.name} is {self.age} years old"
```

### Task 1: Create a Tuxedo Child Class

```python
# TODO: Create your Tuxedo class here
# class Tuxedo(Cat):
#     ...

class TuxedoCat(Cat):
    """A Tuxedo Cat - Hates Wearing Tuxedos"""

    def __init__(self, name, age, has_bow_tie=True):
        super().__init__(name, age)
        self.has_bow_tie = has_bow_tie

    def meow(self):
        return f"{self.name} says meowmeow i hate tuxedos meowmeow!"

    def look_fancy(self):
        return f"{self.name} struts on 2 legs and slicks their whiskers back!"

    def get_info(self):
        return f"{self.name} is {self.age} years old. Bow tie: {self.has_bow_tie}"
```

### Task 2: Create Instances

```python
# Create instances
whiskers = Cat("Whiskers", 4)
mr_mittens = TuxedoCat("Mr. Mittens", 3, True)

print("=== Generic Cat ===")
print(whiskers.meow())
print(whiskers.get_info())

print("\n=== Tuxedo Cat ===")
print(mr_mittens.meow())
print(mr_mittens.get_info())
print(mr_mittens.look_fancy())
print(f"Has bow tie: {mr_mittens.has_bow_tie}")
```
